```python
import requests
# 直接在 URL 夾帶參數
full_url = "https://api-fact-checker.line-apps.com/pub/v1/zhtw/articles/verified?size=9&sort=updatedAt&loc=home"
# get method
response = requests.get(full_url)
print(response.json())  # 輸出網頁內容
```

```python
# 參數不夾帶在URL
base_url = "https://api-fact-checker.line-apps.com/pub/v1/zhtw/articles/verified/"
params = {"size": 9, "sort": "updatedAt", "loc": "home"}
response = requests.get(base_url, params=params)
print(response.json())
```
